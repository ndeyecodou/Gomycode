```python
#importation de bibliotheques
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

dataset = [['Jupe', 'Baskets', 'Écharpe', 'Pantalon', 'Chapeau'],

    ['Lunettes de soleil', 'Jupe', 'Baskets', 'Pantalon', 'Chapeau'],

    ['Robe', 'Sandales', 'Écharpe', 'Pantalon', 'Talons'],

    ['Robe', 'Collier', 'Boucles d\'oreilles', 'Écharpe', 'Chapeau', 'Talons', 'Chapeau'],

   ['Boucles d\'oreilles', 'Jupe', 'Jupe', 'Écharpe', 'Chemise', 'Pantalon']]
#transformation de données en dataframe
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    
df=pd.DataFrame(te_ary, columns=te.columns_)  
df
```

```python

# ajout d'indice pour voir la fréquence de sortie de chaque produit
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
#indice de valeur des données sortie précedemment
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules 
#association des vetement basé sur le metric  de confidence minimale de 70%
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) 
```

```python
#association basée sur la portance
from mlxtend.frequent_patterns import association_rules 
rules= association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
rules
```

```python
#- Les éléments les plus demandés dans l'ensemble de données / Top10


# 1. Gather All Items of Each Transactions into Numpy Array
transaction = []
for i in range(0, data.shape[0]):
    for j in range(0, data.shape[1]):
        transaction.append(data.values[i,j])
transaction = np.array(transaction)
# 2. Transform Them a Pandas DataFrame
df = pd.DataFrame(transaction, columns=["items"]) 

df["incident_count"] = 1 # Put 1 to Each Item For Making Countable Table, to be able to perform Group By

# 3. Delete NaN Items from Dataset
indexNames = df[df['items'] == "nan" ].index
df.drop(indexNames , inplace=True)

# 4. Final Step: Make a New Appropriate Pandas DataFrame
df_table = df.groupby("items").sum().sort_values("incident_count", ascending=False).reset_index()
df_table
```

```python
df_table.tail(10)
```

```python
#pretraitement des données 
#Pour travailler avec l'algorithme apriori, nous devons convertir les transactions en un jeu de données 1/0 ou True/False


# Transform Every Transaction to Seperate List & Gather Them into Numpy Array

transaction = []
for i in range(data.shape[0]):
    transaction.append([str(data.values[i,j]) for j in range(data.shape[1])])
    
transaction = np.array(transaction)
transaction

## Every trnasaction is in on list and combination of list is put in NP array
```

```python
te = TransactionEncoder()
te_ary = te.fit(transaction).transform(transaction)
dataset = pd.DataFrame(te_ary, columns=te.columns_)
dataset
```

```python
first50 = df_table["items"].head(50).values # Select Top50
dataset = dataset.loc[:,first50] # Extract Top50
dataset
```

```python
dataset.columns
# We extracted first 50 column successfully.

## Items - All itesm are my cols 

## TRUE / FALSE  ->> 1 /0 
```

```python
# Convert dataset into 1-0 encoding

def encode_units(x):
    if x == False:
        return 0 
    if x == True:
        return 1
    
dataset = dataset.applymap(encode_units)
dataset.head(10)
```

```python
# Extracting the most frequest itemsets via Mlxtend.

# min support - just to avoid the products which is very lesss occuring ( for 1-2 times in complete dataset )

frequent_itemsets = apriori(dataset, min_support=0.01, use_colnames=True) ## to generate itemset

# The length column has been added to increase ease of filtering.

frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) &
                   (frequent_itemsets['support'] >= 0.04)]
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 3) ].head()
```

```python
# We can create our rules by defining metric and its threshold.

# For a start, 
#      We set our metric as "Lift" to define whether antecedents & consequents are dependent or not.
#      Threshold is selected as "1.2" since it is required to have lift scores above than 1 if there is dependency.


# association_rules - this is used for generating rules now 

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
rules
```

```python
#association basée sur la portance
from mlxtend.frequent_patterns import association_rules 
rules= association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
rules
```

```python
rules["antecedents_length"] = rules["antecedents"].apply(lambda x: len(x))
rules["consequents_length"] = rules["consequents"].apply(lambda x: len(x))
rules.sort_values("lift",ascending=False)
```

```python
#Selon le tableau ci-dessus, nous pouvons facilement dire que la dépendance 
#entre (herb & pepper)et(ground beef) est élevée puisque 
#le score de levage est d'environ 2,5 x le seuil et le score de conviction est supérieur à '1'
```

```python
# Sort values based on confidence

rules.sort_values("confidence",ascending=False)
```

```python
rules[rules["antecedents"].str.contains("ground beef", regex=False) & rules["antecedents_length"] == 1].sort_values("confidence", ascending=False).head(10)
```

```python
***Les scores de confiance, de portance et de soutien de (boeuf haché) et (eau minérale) sont élevés comme prévu. Pour cette raison, le top3 du tableau de tri de confiance était dominé par (boeuf haché)
***
```
