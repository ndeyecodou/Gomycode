```python
#importation du bibliothéque pandas
import pandas as pd
#importation du fichier en csv
data= pd.read_csv('CC GENERAL.csv', encoding='ISO-8859-1')
data
```

```python
data=data.drop(["CUST_ID"], axis=1)

data
```

```python
#verification des valeurs null
data.isnull().sum()
```

```python
data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].mean(),inplace=True)
data['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].mean(),inplace=True)

data
```

```python
#
 #Importation des bibliotheques
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(data)  #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric

agglomerative=pd.DataFrame(clust_labels)
agglomerative

fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (data ['BALANCE'] , data ['PURCHASES'] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clustering")
ax.set_xlabel("Solde restant dans le compte des utilisateur")
ax.set_ylabel("Nombre de transactions d'achat effectuées")
plt.colorbar(scatter)

```

```python
#tracage de la courbe d'endrogramme
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(data, method="complete"))
```

```python
#algorithme de K-means
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=5, random_state=0)  
kmeans.fit(data)
```

```python
#etiquettage
label=pd.DataFrame(kmeans.labels_)
label
#les coordonnées de clusters
kmeans.predict(data)
print(kmeans.cluster_centers_)
```

```python
label
```

```python
plt.scatter(data["BALANCE"][label[0] == 0],          
            data["PURCHASES"][label[0] == 0],s=80,c='magenta',label='Careful')
plt.scatter(data["BALANCE"][label[0] == 1],
           data["PURCHASES"][label[0] == 1],s=80,c='yellow',label='Standard')
plt.scatter(data["BALANCE"][label[0] == 2],
           data["PURCHASES"][label[0] == 2],s=80,c='green',label='Target')
plt.scatter(data["BALANCE"][label[0] == 3], 
           data["PURCHASES"][label[0] == 3],s=80,c='cyan',label='Careless')
plt.scatter(data["BALANCE"][label[0] == 4], 
           data["PURCHASES"][label[0] == 4],s=80,c='burlywood',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title('Cluster')
plt.xlabel('Solde restant dans le compte des utilisateur')
plt.ylabel('Nombre de transactions achat effectuées')
plt.legend()
plt.show()

#on constate avec ce graphique cluster, on a plus d"utilisateur qui font des transaction de maniere standard
```

```python
sum_of_squared_distances=[]
K=range(1,15)
for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(data)
    sum_of_squared_distances.append(km.inertia_)

plt.plot(K,sum_of_squared_distances,'bx-')
plt.xlabel('k')
plt.ylabel('sum_of_squared_distances')
plt.title('La valeur de K optimal')
plt.show()
```
