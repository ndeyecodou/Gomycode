```python
#histogramme des differente colomnes
df.hist(figsize=(30,20))
plt.show()   
# Divisez votre ensemble de données en un ensemble d'apprentissage et un ensemble de test

#importation des librairies
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn import metrics
from sklearn.preprocessing import PolynomialFeatures 
#définition des valeurs de X et Y
#Y: le prix
Y=df.iloc[:,[0]].values
#X: definit les autres  valeurs
X=df.iloc[:,1:].values


```

```python
#Diviser le dataset
X_train, X_test, Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
#construction modéle
regression= LinearRegression()
regression.fit(X_train,Y_train)

regression.coef_
```

```python
#faire des predictions
Y_pred=regression.predict(X_test)


print("MSE: ", metrics.mean_squared_error(Y_test, Y_pred)**2)
print("R : ",r2_score(Y_test,Y_pred))
```
